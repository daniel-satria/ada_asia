# Feature Engineer

```python
X = master.drop(['tcm_campaign_id', 'tcm_flag_summary',
                'tcm_count_retry_change_status', 'tcm_flag_get_report',
                'tcm_total_number', 'tcm_total_sent', 'tcm_id', 'tcgc_status'], axis=1)
y = master['tcgc_status']
```

```python
X.head()
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
        X, y, test_size=0.33, random_state=42)
```

```python
# slicing data by category for data test
X_test_numerical = X_test[['tcgc_budget', 'tcm_success']]
X_test_ordinal = X_test[['tcgc_type', 'tcgc_flag_lock', 'tcm_status_campaign',
                    'tcm_channel','feature_tf_id', 'tcm_inventory_id',
                    'tcm_external_entity_id']]

# concatting the final X_Train
X_test_final = pd.concat([X_test_numerical, X_test_ordinal], axis=1)

X_test_final.head()
```

```python
from sklearn.preprocessing import MinMaxScaler
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OrdinalEncoder
from sklearn.ensemble import RandomForestClassifier 

# slicing data by category
X_train_numerical = X_train[['tcgc_budget', 'tcm_success']]
X_train_ordinal = X_train[['tcgc_type', 'tcgc_flag_lock', 'tcm_status_campaign',
                    'tcm_channel','feature_tf_id', 'tcm_inventory_id',
                    'tcm_external_entity_id']]

# concatting the final X_Train
X_train_final = pd.concat([X_train_numerical,X_train_ordinal], axis=1)

# initiate scaler
scaler = MinMaxScaler()
encoder = OrdinalEncoder(handle_unknown='use_encoded_value', unknown_value=-2)

# get the values from column
num_col = X_train_numerical.columns.values
ordinal_col = X_train_ordinal.columns.values

numerical_pipe = Pipeline([
    ('numerical_encode', MinMaxScaler())
    ])

ordinal_pipe = Pipeline([
    ('ordinal_encoder', OrdinalEncoder(handle_unknown='use_encoded_value', unknown_value=-2))
    ])

model_pipe = Pipeline([
    ('model_rf', RandomForestClassifier(max_depth=5, random_state=100))
    ])

preprocess = ColumnTransformer([
    ('num', numerical_pipe, num_col),
    ('ord', ordinal_pipe, ordinal_col)
    ])


pipe = Pipeline(steps=[('preprocess', preprocess), ('random_forest', model_pipe)])

pipe.fit(X_train_final, y_train)
```

## Evaluation 

```python
# getting the predicted value of y_train
y_train_pred = pipe.predict(X_train_final)

# getting the predicted value of y_test
y_test_pred = pipe.predict(X_test_final)
```

```python
from sklearn.metrics import f1_score

print("Train score ", f1_score(y_train, y_train_pred, average='micro'))
print("Test score ", f1_score(y_test, y_test_pred, average='micro'))
```
